```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

$$\begin{cases}
\nabla \times (\nabla \times \mathbf{u}) = \mathbf{0} &  \\
\nabla \cdot \mathbf{u} = 0 & 
\end{cases}$$

$$\begin{aligned}
u_x(r, \theta) = \frac{2}{3} r^{-\frac{1}{3}} \cos\left(\frac{1}{3}\theta\right) \\[8pt]
u_y(r, \theta) = -\frac{2}{3} r^{-\frac{1}{3}} \sin\left(\frac{1}{3}\theta\right)
\end{aligned}$$

```python
def get_maxwell_exact(x, y):
    # 极坐标
    r = np.sqrt(x**2 + y**2) + 1e-8
    theta = np.arctan2(y, x)
    theta = np.where(theta < 0, theta + 2*np.pi, theta)
    
    alpha = 2/3.0
    
    dPhi_dr = alpha * (r**(alpha - 1)) * np.sin(alpha * theta)
    dPhi_dtheta = alpha * (r**alpha) * np.cos(alpha * theta)
    
    cos_t = x / r
    sin_t = y / r
    
    dPhi_dx = cos_t * dPhi_dr - (sin_t / r) * dPhi_dtheta
    dPhi_dy = sin_t * dPhi_dr + (cos_t / r) * dPhi_dtheta
    
    u_x = dPhi_dy
    u_y = -dPhi_dx
    
    return u_x, u_y

def get_maxwell_exact_torch(x_tensor):
    x = x_tensor[:, 0:1]
    y = x_tensor[:, 1:2]
    
    r = torch.sqrt(x**2 + y**2) + 1e-6
    theta = torch.atan2(y, x)
    theta = torch.where(theta < 0, theta + 2*np.pi, theta)
    
    alpha = 2/3.0
    
    dPhi_dr = alpha * (r**(alpha - 1)) * torch.sin(alpha * theta)
    dPhi_dtheta = alpha * (r**alpha) * torch.cos(alpha * theta)
    
    cos_t = x / r
    sin_t = y / r
    
    dPhi_dx = cos_t * dPhi_dr - (sin_t / r) * dPhi_dtheta
    dPhi_dy = sin_t * dPhi_dr + (cos_t / r) * dPhi_dtheta
    
    return torch.cat([dPhi_dy, -dPhi_dx], dim=1)
```

```python
class CubicReLU(nn.Module):
    def forward(self, x):
        return torch.pow(torch.relu(x), 3)

class Sine(nn.Module):
    def forward(self, x):
        return torch.sin(x)

class VectorPINN(nn.Module):
    def __init__(self, activation_name='tanh'):
        super().__init__()
        
        if activation_name == 'tanh':
            self.act = nn.Tanh()
        elif activation_name == 'relu':
            self.act = nn.ReLU()
        elif activation_name == 'relu3':
            self.act = CubicReLU()
        elif activation_name == 'rational':
            self.act = None # 特殊处理
            self.act_fn = lambda x: x / (1 + torch.abs(x))
        else:
            self.act = nn.Tanh()
            
        self.activation_name = activation_name
        
        # 定义层
        self.fc1 = nn.Linear(2, 64)
        self.fc2 = nn.Linear(64, 2048)
        self.fc3 = nn.Linear(2048, 64)
        self.fc4 = nn.Linear(64, 2)

        
    def forward(self, x):
        if self.activation_name == 'rational':
            h = self.fc1(x)
            h = self.act_fn(h)
            h = self.fc2(h)
            h = self.act_fn(h)
            h = self.fc3(h)
            h = self.act_fn(h)
            h = self.fc4(h)
            return h
        else:
            h = self.act(self.fc1(x))
            h = self.act(self.fc2(h))
            h = self.act(self.fc3(h))
            return self.fc4(h)
```

```python
def maxwell_residual(model, x_in):
    x_in.requires_grad = True
    u = model(x_in)
    u_x = u[:, 0:1]
    u_y = u[:, 1:2]
    grad_ux = torch.autograd.grad(u_x, x_in, torch.ones_like(u_x), create_graph=True)[0]
    grad_uy = torch.autograd.grad(u_y, x_in, torch.ones_like(u_y), create_graph=True)[0]
    
    d_ux_dx = grad_ux[:, 0:1]
    d_ux_dy = grad_ux[:, 1:2]
    d_uy_dx = grad_uy[:, 0:1]
    d_uy_dy = grad_uy[:, 1:2]
    scalar_curl = d_uy_dx - d_ux_dy
    
    #  (d(curl)/dy, -d(curl)/dx)
    grad_curl = torch.autograd.grad(scalar_curl, x_in, torch.ones_like(scalar_curl), create_graph=True)[0]
    d_curl_dx = grad_curl[:, 0:1]
    d_curl_dy = grad_curl[:, 1:2]
    
    res_x = d_curl_dy 
    res_y = -d_curl_dx
    
    # Div u = 0 
    res_div = d_ux_dx + d_uy_dy
    
    return res_x, res_y, res_div
```

```python
#内部
pts = np.random.rand(12000, 2) * 2 - 1 
mask = ~((pts[:, 0] > 0) & (pts[:, 1] < 0))
x_col = torch.tensor(pts[mask], dtype=torch.float32).to(device)

# 边界点
n1_x = torch.rand(300, 1)
x_bc1 = torch.cat([-n1_x, -torch.ones_like(n1_x)], dim=1)

n2_y = torch.rand(600, 1) * 2 - 1
x_bc2 = torch.cat([-torch.ones_like(n2_y), n2_y], dim=1)

n3_x = torch.rand(600, 1) * 2 - 1
x_bc3 = torch.cat([n3_x, torch.ones_like(n3_x)], dim=1)

n4_y = torch.rand(300, 1)
x_bc4 = torch.cat([torch.ones_like(n4_y), n4_y], dim=1)

n5_x = torch.rand(300, 1)
x_bc5 = torch.cat([n5_x, torch.zeros_like(n5_x)], dim=1)

n6_y = torch.rand(300, 1)
x_bc6 = torch.cat([torch.zeros_like(n6_y), -n6_y], dim=1)

x_bc = torch.cat([x_bc1, x_bc2, x_bc3, x_bc4, x_bc5, x_bc6], dim=0).float().to(device)

u_bc_true = get_maxwell_exact_torch(x_bc).to(device)
```

```python
ACT_TYPE = 'relu3' 

model = VectorPINN(ACT_TYPE).to(device)
optimizer = optim.Adam(model.parameters(), lr=5e-5)
print(f"--- Training Maxwell PINN ({ACT_TYPE}) on {device} ---")

loss_history = []
# 训练
for epoch in range(20001):
    optimizer.zero_grad()
    
    # PDE Loss
    rx, ry, rdiv = maxwell_residual(model, x_col)
    loss_pde = torch.mean(rx**2 + ry**2) + torch.mean(rdiv**2)
    
    # BC Loss
    u_pred_bc = model(x_bc)
    loss_bc = torch.mean((u_pred_bc - u_bc_true)**2)
    
    # Total Loss
    loss = loss_pde + 10 * loss_bc
    
    loss.backward()
    optimizer.step()
    
    loss_history.append(loss.item())
    
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss {loss.item():.6f} (PDE: {loss_pde.item():.6f}, BC: {loss_bc.item():.6f})")
```

```python
for epoch in range(20000,30001):
    optimizer.zero_grad()
    
    # PDE Loss
    rx, ry, rdiv = maxwell_residual(model, x_col)
    loss_pde = torch.mean(rx**2 + ry**2) + torch.mean(rdiv**2)
    
    # BC Loss
    u_pred_bc = model(x_bc)
    loss_bc = torch.mean((u_pred_bc - u_bc_true)**2)
    
    # Total LossQ
    loss = loss_pde + 10 * loss_bc
    
    loss.backward()
    optimizer.step()
    
    loss_history.append(loss.item())
    
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss {loss.item():.6f} (PDE: {loss_pde.item():.6f}, BC: {loss_bc.item():.6f})")
```

```python
for epoch in range(30000,40001):
    optimizer.zero_grad()
    
    # PDE Loss
    rx, ry, rdiv = maxwell_residual(model, x_col)
    loss_pde = torch.mean(rx**2 + ry**2) + torch.mean(rdiv**2)
    
    # BC Loss
    u_pred_bc = model(x_bc)
    loss_bc = torch.mean((u_pred_bc - u_bc_true)**2)
    
    # Total LossQ
    loss = loss_pde + 10 * loss_bc
    
    loss.backward()
    optimizer.step()
    
    loss_history.append(loss.item())
    
    if epoch % 500 == 0:
        print(f"Epoch {epoch}: Loss {loss.item():.6f} (PDE: {loss_pde.item():.6f}, BC: {loss_bc.item():.6f})")
```

```python
def evaluate_grid(x_range, y_range, n_grid, model, device):
    x = np.linspace(x_range[0], x_range[1], n_grid)
    y = np.linspace(y_range[0], y_range[1], n_grid)
    X, Y = np.meshgrid(x, y)
    XY = np.stack([X.ravel(), Y.ravel()], 1)
    XY_torch = torch.tensor(XY, dtype=torch.float32).to(device)
    
    # L-shape 
    geom_mask = (X > 0) & (Y < 0)
    
    with torch.no_grad():
        U_vec = model(XY_torch).cpu().numpy()
        U_mag_pred = np.sqrt(U_vec[:,0]**2 + U_vec[:,1]**2).reshape(n_grid, n_grid)
        
    ux_true, uy_true = get_maxwell_exact(X, Y)
    U_mag_true = np.sqrt(ux_true**2 + uy_true**2)
    
    clip_val = 100.0
    U_mag_true = np.clip(U_mag_true, 0, clip_val)
    U_mag_pred = np.clip(U_mag_pred, 0, clip_val)

    err = np.abs(U_mag_pred - U_mag_true)
    
    U_mag_true[geom_mask] = np.nan
    U_mag_pred[geom_mask] = np.nan
    err[geom_mask] = np.nan
    
    return X, Y, U_mag_true, U_mag_pred, err
```

```python
n_grid = 120
X_g, Y_g, True_g, Pred_g, Err_g = evaluate_grid((-1, 1), (-1, 1), n_grid, model, device)
X_m, Y_m, True_m, Pred_m, Err_m = evaluate_grid((-0.1, 0.1), (-0.1, 0.1), n_grid, model, device)
X_s, Y_s, True_s, Pred_s, Err_s = evaluate_grid((-0.01, 0.01), (-0.01, 0.01), n_grid, model, device)
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 1. 全局真值
im0 = axes[0, 0].contourf(X_g, Y_g, True_g, levels=50, cmap='jet')
axes[0, 0].set_title("1. Global Exact Field |u|")
fig.colorbar(im0, ax=axes[0, 0])

# 2. 全局预测
im1 = axes[0, 1].contourf(X_g, Y_g, Pred_g, levels=50, cmap='jet')
axes[0, 1].set_title(f"2. Global PINN ({ACT_TYPE})")
fig.colorbar(im1, ax=axes[0, 1])

# 3. 全局误差
im2 = axes[0, 2].contourf(X_g, Y_g, Err_g, levels=50, cmap='inferno')
axes[0, 2].set_title("3. Global Error")
fig.colorbar(im2, ax=axes[0, 2])

# 4. 真值 (Zoom 0.1)
im3 = axes[1, 0].contourf(X_m, Y_m, True_m, levels=50, cmap='jet')
axes[1, 0].set_title("4. Global Exact Field |u| (Zoom 0.1)")
fig.colorbar(im0, ax=axes[1, 0])

# 5. 预测 (Zoom 0.1) 
im4 = axes[1, 1].contourf(X_m, Y_m, Pred_m, levels=50, cmap='jet')
axes[1, 1].set_title(f"5. PINN Prediction (Zoom 0.1)")
fig.colorbar(im3, ax=axes[1, 1])

# 6. 误差 (Zoom 0.1)
im5 = axes[1, 2].contourf(X_m, Y_m, Err_m, levels=50, cmap='inferno')
axes[1, 2].set_title("6. Error (Zoom 0.1)")
fig.colorbar(im5, ax=axes[1, 2])

plt.suptitle(f"Maxwell Singularity Test: ACT={ACT_TYPE}, Epochs=40000", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 0.1 范围下 u的x分量
n_grid_ux = 150
x = np.linspace(-0.1, 0.1, n_grid_ux)
y = np.linspace(-0.1, 0.1, n_grid_ux)
X, Y = np.meshgrid(x, y)
XY = np.stack([X.ravel(), Y.ravel()], axis=1)

XY_torch = torch.tensor(XY, dtype=torch.float32).to(device)
with torch.no_grad():
    U_pred = model(XY_torch).cpu().numpy()

u_x_pred = U_pred[:, 0].reshape(n_grid_ux, n_grid_ux)
u_x_true, _ = get_maxwell_exact(X, Y)

mask = (X > 0) & (Y < 0)
u_x_true[mask] = np.nan
u_x_pred[mask] = np.nan
u_x_err = np.abs(u_x_pred - u_x_true)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

im0 = axes[0].contourf(X, Y, u_x_true, levels=50, cmap='jet')
axes[0].set_title(r"Exact $u_x$ (Zoom 0.1)")
fig.colorbar(im0, ax=axes[0])

im1 = axes[1].contourf(X, Y, u_x_pred, levels=50, cmap='jet')
axes[1].set_title(r"PINN $u_x$ (Zoom 0.1)")
fig.colorbar(im1, ax=axes[1])

im2 = axes[2].contourf(X, Y, u_x_err, levels=50, cmap='inferno')
axes[2].set_title(r"$|u_x^{pred} - u_x^{exact}|$")
fig.colorbar(im2, ax=axes[2])

plt.suptitle(r"$u_x$ near re-entrant corner ($[-0.1,0.1]^2$)", fontsize=14)
plt.tight_layout()
plt.show()
```
